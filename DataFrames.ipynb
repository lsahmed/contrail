## Pandas Series
A series is a one-dimensional labeled array capable of holding any data type. The axis labels are correctively called the index.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
pd.Series([10,20,30],index=['a','b','c'])
```

## Creating a DataFrame

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns=columns)
```

```python
df2
```

## Selection and Indexing of Columns

```python
df2[['Name','City']]
```

## Creating a new column

```python
df2['Designation'] = ['Engineer','Fashion Designer','Ethical Hacker','Doctor']
df2
```

## Removing Column

```python
df2.drop('Designation', axis = 1, inplace = True)
```

```python
df2
```

## Selecting Rows

```python
df2
```

```python
df2.loc[[0,1]]
```

```python
df2.iloc[3]
```

## Selecting subsets of Rows and columns

```python
df2.loc[[0,1]][["City","Salary"]]
```

```python
df2.loc[[2,3]][["Name","Age"]]
```

## Conditional Selection

```python
# I only want to see those people whose age is above 30
```

```python
df2[df2["Age"]>30]
```

```python
df2[(df2["Age"]>30) & (df2["City"] == "Paris")]
```
